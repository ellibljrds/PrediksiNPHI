Langkah berikutnya yang saya lakukan adalah memasukkan zona-zona sesuai dengan sumur dan kedalaman.

```python
def get_layer(data):
    depths = lookup[data['Well']]
    if data['Depth'] < list(depths.keys())[-1]:
        return 'A'
    elif data['Depth'] > list(depths.keys())[0]:
        return 'B'
    else:
        for lim, layer in depths.items():
            if data['Depth'] >= lim:
                return layer
            
lookup = {}
for (k1,k2),v in df2.set_index(['Well','Depth'])['Layer'][::-1].to_dict().items():
    lookup.setdefault(k1, {}).update({k2:v})

df['Layer'] = df.apply(get_layer, axis=1)
df.head()
```

Langkah berikutnya yang saya lakukan adalah menghapus outlier beserta data dengan nilai yang mustahil atau tidak mungkin secara fisika. Biasanya, nilai ini keluar karena kondisi lubang bor yang tidak ideal.

```python
df = df.replace(-999.25, np.NaN)
df['ILD'] = np.where(df['ILD'] > 10000, np.NaN, df['ILD'])
df['GR'] = np.where(df['GR'] > 250, np.NaN, df['GR'])
df['NPHI'] = np.where(df['NPHI'] > 0.6, np.NaN, df['NPHI'])
df['NPHI'] = np.where(df['NPHI'] < 0, np.NaN, df['NPHI'])
df['RHOB'] = np.where(df['RHOB'] < 0, np.NaN, df['RHOB'])
df['RHOB'] = np.where(df['RHOB'] > 3, np.NaN, df['RHOB'])
df['DT'] = np.where(df['DT'] < 0, np.NaN, df['DT'])
df.head()
```

```python
dfax = pd.read_csv('Porto tarbertz.csv')
df_prop = pd.read_csv('Porto tarbertz2.csv')
dfax = dfax.rename(columns={'Well identifier':'Well'})
df_prop = df_prop.rename(columns={'Zone log': 'Surface',
                                  'Porosity': 'PHIE',
                                  'Continuous': 'NTG'})
dfax.head()
```

```python
bot  = dfax[dfax['Surface'].str.contains("Bot")]
bot.drop('Prod', axis = 1, inplace = True)
bot['Surface'] = bot['Surface'].str[4:] ## Di sini saya menghapus 4 string pertama (karena semua format penulisannya sama) di kolom surface untuk menghapus Bot_.
bot = bot.rename(columns={'MD': 'MD_Bot'})
bot.head()
```

```python
dfax = dfax[~dfax.Surface.str.contains('|'.join(['Bot|bot']))]
dfax = dfax.replace('Top_', '', regex=True)
dfax.reset_index(drop=True, inplace=True)
dfax.head()
```

```python
dfax1 = pd.merge(dfax, bot, how = "left", on=["Well", "Surface"])
dfax1.head()
```

```python
dfax1['Thickness'] = dfax1['MD_Bot'] - dfax1['MD']
dfax1 = dfax1[dfax1['Thickness'].notna()]
dfax1.head(6)
```

```python
c_maxes = dfax1.groupby(['Well', 'Surface']).Thickness.transform(max)
dfax1 = dfax1.loc[dfax1.Thickness == c_maxes]
dfax1.head()
```

```python
df_prop = df_prop.rename(columns={'Zone log': 'Surface',
                                  'Porosity': 'PHIE',
                                  'Continuous': 'NTG'})
df_prop['Surface'] = (df_prop.Surface.str.replace('Zone ', ''))
df_prop['Surface'] = (df_prop.Surface.str.replace('Top_', ''))
df_prop.head()
```

```python
dfax2 = pd.merge(dfax1, df_prop, how = "left", on=["Well", "Surface"])
dfax2.head()
```

```python
dfax2.dropna(subset = ['PHIE', 'Perm', 'NTG'], inplace=True)
dfax2.reset_index(drop=True, inplace=True)
dfax2['Net_Sand'] = dfax2['Thickness'] * dfax2['NTG']
dfax2.head()
```
